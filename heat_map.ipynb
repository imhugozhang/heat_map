```python
import gmaps.datasets
import folium
from folium import plugins
import pandas as pd
```

```python
#obtain data
gmaps.configure(api_key="AI...")    #it is ok not to type in a real API Key in this case
#locations of taxis in San Francisco in a particular time
locations = gmaps.datasets.load_dataset("taxi_rides")
print(locations)
```

```python
#draw the map
m = folium.Map([37.778926, -122.416269], zoom_start=13)
for location in locations:
    folium.CircleMarker([location[0], location[1]],
                        radius=1,
                        ).add_to(m)
#add heat map layer to the map
m.add_child(folium.plugins.HeatMap(locations, radius=30))
```
